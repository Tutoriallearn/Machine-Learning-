```python
import pandas as pd


```

```python
data = {
    "Name": ["A", "B", "C"],
    "Age": [25, 40, 30],
    "Salary": [50000, 80000, 60000],
    "Join_Date": ["2020-01-15", "2018-07-10", "2021-03-20"]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
#Salary_per_Age New Column Create

df["Salary_per_Age"] = df["Salary"] / df["Age"]
```

```python
df
```

```python
#Join_Date object to datetime convert
df["Join_Date"] = pd.to_datetime(df["Join_Date"])

#Join_Year New Column Create
df["Join_Year"] = df["Join_Date"].dt.year

#Join_Month New Column Create
df["Join_Month"] = df["Join_Date"].dt.month

```

```python
df
```

```python
#Age_Group New Column Create and convert to group
df["Age_Group"] = pd.cut(df["Age"],
                        bins=[0, 30, 50],
                        labels=["Young", "Adult"])

```

```python
df
```

```python
#Age_Salary New Column Create multiplication of Age and Salary
df["Age_Salary"] = df["Age"] * df["Salary"]
```

```python
df
```

```python
#High_Salary New Column Create salary greater than 60000 give boolean value
df["High_Salary"] = df["Salary"] > 60000
```
