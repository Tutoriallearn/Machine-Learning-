```python
import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/home/diabetes.csv')
```

```python
df
```

```python
sns.kdeplot(df['Glucose'])
```

```python
sns.kdeplot(df['BMI'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Outcome', axis=1),
                                                    df['Outcome'],
                                                    test_size=0.3,
                                                    random_state=0)

X_train.shape, X_test.shape

X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# fit
scaler.fit(X_train)

# transform
X_train_Normalized = scaler.transform(X_train)
X_test_Normalized = scaler.transform(X_test)
```

```python
X_train_Normalized = pd.DataFrame(X_train_Normalized, columns=X_train.columns)
X_test_Normalized = pd.DataFrame(X_test_Normalized, columns=X_test.columns)
```

```python
np.round(X_train.describe(), 1)
```

```python
np.round(X_train_Normalized.describe(), 1)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Glucose'], X_train['Insulin'],c=y_train)
ax1.set_title("Before Normalized")
ax2.scatter(X_train_Normalized['Glucose'], X_train_Normalized['Insulin'],c=y_train)
ax2.set_title("After  Normalized")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# Before Normalized
ax1.set_title('Before Normalized')
sns.kdeplot(X_train['Glucose'], ax=ax1)
sns.kdeplot(X_train['Insulin'], ax=ax1)

# After Normalized
ax2.set_title('After Normalized')
sns.kdeplot(X_train_Normalized['Glucose'], ax=ax2)
sns.kdeplot(X_train_Normalized['Insulin'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# Before Normalized
ax1.set_title('Glucose Distribution Before Normalized')
sns.kdeplot(X_train['Glucose'], ax=ax1)

# After Normalized
ax2.set_title('Glucose Distribution After Normalized')
sns.kdeplot(X_train_Normalized['Glucose'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# Before Normalized
ax1.set_title('Insulin Distribution Before Normalized')
sns.kdeplot(X_train['Insulin'], ax=ax1)

# After Normalized
ax2.set_title('Insulin Distribution After Normalized')
sns.kdeplot(X_train_Normalized['Insulin'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# Before Normalized
ax1.set_title('Age Distribution Before Normalized')
sns.kdeplot(X_train['Age'], ax=ax1)

# After Normalized
ax2.set_title('Age Distribution After Normalized')
sns.kdeplot(X_train_Normalized['Age'], ax=ax2)
plt.show()
```
