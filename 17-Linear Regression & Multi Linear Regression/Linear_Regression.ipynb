```python
# Step 0: Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

```

 ## Step 1: Create dataset (Experience vs Salary)

```python

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)   # input
y = np.array([30000, 35000, 40000, 45000, 50000])  # output

```

 ### Step 2: Create model

```python

model = LinearRegression()

```

### Step 3: Train model (learn m and b)

```python

model.fit(X, y)

```

 ### Step 4: Make prediction

```python

new_exp = np.array([[6]])
predicted_salary = model.predict(new_exp)

```

###  Step 5: Print results

```python

print("Slope (m):", model.coef_)
print("Intercept (b):", model.intercept_)
print("Predicted salary for 6 years:", predicted_salary)

```

### Step 6: Visualization

```python

plt.scatter(X, y, label="Actual Data")
plt.plot(X, model.predict(X), label="Regression Line")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
