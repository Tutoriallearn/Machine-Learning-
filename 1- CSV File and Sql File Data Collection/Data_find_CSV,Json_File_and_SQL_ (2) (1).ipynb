###2. Opening a csv file from an URL

```python

import requests
from io import StringIO

url = "https://raw.githubusercontent.com/Tutoriallearn/Pandas-For-Data-Science-and-AI-ML/main/orders.csv"

headers = {"User-Agent": "Mozilla/5.0"}
req = requests.get(url, headers=headers)

data = StringIO(req.text)

df = pd.read_csv(data)

print(df)
```

###9. Skip bad lines

```python
df = pd.read_csv(
    '/home/connecticut_real_estate_2026.csv',
    sep=';',
    encoding='latin-1',
    on_bad_lines='skip'
)

print(df)
```
