```python
import pandas as pd
```

```python
data = {
    "Customer_Name": ["John Doe", "Alice Smith", "Bob Johnson"],
    "Order_ID": ["ORD-12345", "ORD-67890", "ORD-54321"],
    "Date": ["2026-04-14", "2026-05-20", "2026-06-10"]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# Customer_Name  convert to First_name And Last_Name two Diffrent Column
df[["First_Name", "Last_Name"]] = df["Customer_Name"].str.split(" ", expand=True)

```

```python
df
```

```python
# Order_ID  convert to Order_Type and Order_Number two Diffrent Column
df[["Order_Type", "Order_Number"]] = df["Order_ID"].str.split("-", expand=True)

```

```python
df
```

```python

# Date convert to Year,Month,Day three Diffrent Column
df["Date"] = pd.to_datetime(df["Date"])

df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day

```

```python
df
```

```python
#Drop Existing Column After Splitting
df.drop(columns=["Customer_Name", "Order_ID", "Date"], inplace=True)
```
