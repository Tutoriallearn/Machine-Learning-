###BiVariant Analysis

###1. Scatterplot (Numerical - Numerical)

```python
df2 = pd.read_csv('/home/job_salary_prediction_dataset.csv')
```

```python
df3=df2.head(100)
```

```python
df3
```

```python
sns.scatterplot(x=df3['experience_years'],y=df3['skills_count'],hue=df3['education_level'],
                style=df3['company_size'],size=df3['certifications'])
```

###2. Bar Plot (Numerical - Categorical)

```python
sns.barplot(data= df3,x=df3['company_size'],y=df3['experience_years'],hue=df3['education_level'])
```

###3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(data= df3,x=df3['company_size'],y=df3['experience_years'],hue=df3['education_level'])
```

###4. Distplot (Numerical - Categorical)

```python
sns.distplot(df3[df3['certifications']==0]['experience_years'],hist=False)
```

```python
sns.distplot(df3[df3['certifications']==1]['experience_years'],hist=False)
```

###5. HeatMap (Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(df3['education_level'],df3['skills_count']))
```

###6. ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(df3['education_level'],df3['skills_count']))
```

###7. Pairplot

```python
sns.pairplot(df3,hue='industry')
```

### 8. Lineplot (Numerical - Numerical)

```python
sns.lineplot(data=df3,x='experience_years',y='skills_count')
```
