```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

data = {
    "City": ["Delhi", "Mumbai", np.nan, "Chennai", "Delhi"],
    "Age": [25, np.nan, 30, 35, 40],
    "Salary": [50000, 60000, 55000, np.nan, 65000]
}

df = pd.DataFrame(data)
print(df)
```

###Using Pandas

```python

num_imputer = SimpleImputer(strategy="mean")

df["Age"] = num_imputer.fit_transform(df[["Age"]])
df["Salary"] = num_imputer.fit_transform(df[["Salary"]])
```

```python
cat_imputer = SimpleImputer(strategy="most_frequent")

df["City"] = cat_imputer.fit_transform(df[["City"]])[:, 0]
```

###Using SkLearn

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="constant", fill_value="Unknown")

df_imputed = imputer.fit_transform(df)

df_imputed = pd.DataFrame(df_imputed, columns=df.columns)

print(df_imputed)
```

```python
encoder = OneHotEncoder( handle_unknown="ignore",
    sparse_output=False)

encoded = encoder.fit_transform(df[["City"]])

encoded_df = pd.DataFrame(
    encoded,
    columns=encoder.get_feature_names_out(["City"])
)

df_final = pd.concat([df.drop("City", axis=1), encoded_df], axis=1)

print(df_final)
```

```python
from sklearn.compose import ColumnTransformer

num_cols = ["Age", "Salary"]
cat_cols = ["City"]

preprocessor = ColumnTransformer([
    ("num", SimpleImputer(strategy="mean"), num_cols),
    ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols)
])

df_processed = preprocessor.fit_transform(df)

print(df_processed)
```
