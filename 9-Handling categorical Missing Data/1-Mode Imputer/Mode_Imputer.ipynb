```python
import pandas as pd
import numpy as np

data = {
    'Gender': ['Male', 'Female', np.nan, 'Female', 'Male', np.nan, 'Male'],
    'City': ['Delhi', 'Mumbai', 'Delhi', np.nan, 'Delhi', 'Mumbai', np.nan],
    'Purchased': ['Yes', 'No', 'Yes', 'Yes', np.nan, 'No', 'Yes']
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.isnull().mean()*100
```

```python
df['Gender'].value_counts().plot(kind='bar')
```

```python
df['City'].value_counts().plot(kind='bar')
```

```python
df['Purchased'].value_counts().plot(kind='bar')
```

```python
df.mode()
```

```python
df_mode = df.copy()

df_mode['Gender'] = df_mode['Gender'].fillna(df_mode['Gender'].mode()[0])
df_mode['City'] = df_mode['City'].fillna(df_mode['City'].mode()[0])
df_mode['Purchased'] = df_mode['Purchased'].fillna(df_mode['Purchased'].mode()[0])
```

```python
df_mode
```

```python
df_mode.isnull().sum()
```

###Sk Learn Use

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='most_frequent')

df_sklearn = pd.DataFrame(
    imputer.fit_transform(df),
    columns=df.columns
)

```

```python
df_sklearn
```
