```python
import pandas as pd
import numpy as np
from scipy import stats
```

###Create sample data:

```python


data = {
    'marks': [40, 42, 45, 47, 50, 52, 100]  # 100 is an outlier
}
df = pd.DataFrame(data)
print(df)
```

```python
df['marks'].describe()
```

###Calculate Z-score:

```python
mean = df['marks'].mean()
std = df['marks'].std()

df['z_score_manual'] = (df['marks'] - mean) / std

print(df)
```

```python
df['z_score'] = stats.zscore(df['marks'])
```

```python
df['z_score']
```

###Detect outliers:

```python


outliers = df[np.abs(df['z_score']) > 3]
print("Outliers:\n", outliers)
```

###Remove outliers:

```python


df_clean = df[np.abs(df['z_score']) <= 3]
print("Clean Data:\n", df_clean)
```
